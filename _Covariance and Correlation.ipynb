# Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import dataset

```python
dataset = pd.read_excel("C:/Users/ashuj/OneDrive/Desktop/Mobile Sales Data.xlsx")
```

```python
dataset.head(10)
```

# Showing the information about the dataset

```python
dataset.info()
```

# Calculating Correlation and Covariance of the dataset

```python
data_corr = dataset.select_dtypes("int64","float64").corr()
data_corr
```

```python
data_cov = dataset.select_dtypes("int64","float64").cov()
data_cov
```

# Showing Correlation and Covariance in heatmap graph with it's values

```python
plt.figure(figsize=(7,5))
sns.heatmap(data_corr,annot=True)
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.heatmap(data_cov,annot=True)
plt.show()
```
