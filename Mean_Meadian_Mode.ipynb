# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing data 

```python
dataset = pd.read_excel("C:/Users/ashuj/OneDrive/Desktop/Mobile Sales Data.xlsx")
```

# Showing data of starting 10 rows

# Finding Mean of the Customer Age from the dataset

```python
dataset["Customer Age"].mean()
```

```python
me = np.mean(dataset["Customer Age"])
me
```

# Finding Median of the Customer Age column from the dataset

```python
md = np.median(dataset["Customer Age"])
md
```

# Finding Mode of the Customer Age column from the dataset 

```python
mo =dataset["Customer Age"].mode()[0]
mo
```

# Count the Customer Age in the dataset 

```python
dataset["Customer Age"].value_counts().head(5)
```

# Checking the null value are present or not in the dataset 

```python
dataset.isnull().sum()
```

# Plotting the graph using Seaborn of the Mean, Median and Mode of the dataset 

```python
sns.histplot(x="Customer Age",data=dataset,bins=[i for i in range (0,61,10)])
plt.plot([me for i in range (0,1000)],[i for i in range (0,1000)],c="yellow",label="mean")
plt.plot([md for i in range (0,1000)],[i for i in range (0,1000)],c="blue",label="median")
plt.plot([mo for i in range (0,1000)],[i for i in range (0,1000)],c="red",label="mode")
plt.legend()
plt.show()
```
